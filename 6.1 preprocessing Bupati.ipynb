```python
import os
import pandas as pd
import numpy as np
from glob import glob
from tqdm import tqdm
```

```python
list_files = glob('./Bupati/*.csv')
len(list_files)
```

```python
dfs = []
for file in list_files:
    child = pd.read_csv(file)
    dfs.append(child)
```

```python
df = pd.concat(dfs, ignore_index=True)
df
```

```python
df.isnull().mean()*100
```

```python
df["nama"] = df["nama"].str.replace("Petahana", " ")
```

```python
df["nama"] = [x[0] for x in df["nama"].str.split("sejak")]
```

```python
df["nama"] = [x[1] if len(x) > 1 else x[0] for x in df["nama"].str.split(".jpg")]
df["nama"] = [x[1] if len(x) > 1 else x[0] for x in df["nama"].str.split(".JPG")]
```

```python
df = df[["nama", "alamat", "daerah_sumber"]]
df
```

```python
df.to_excel("./Bupati/all/list_bupati.xlsx", index=False)
```
