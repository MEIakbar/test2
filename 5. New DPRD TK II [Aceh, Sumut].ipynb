# Function

```python
def get_url(url):
    headers = {
         'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.142 Safari/537.36'
     }
    r = requests.get(url, headers=headers)  # Using the custom headers we defined above
    soup = BeautifulSoup(r.content, 'lxml') 
    return soup
```

```python
li = []

for url in tqdm(urls):
    df = pd.read_html(url)[0]
    print(df.shape)
    li.append(df)

df = pd.concat(li)
df
```

```python
new_header = df.iloc[0] #grab the first row for the header
df = df[1:] #take the data less the header row
df.columns = new_header #set the header row as the df header
df
```

```python
df = df.rename(columns={"NAMA" : "nama"})
df["alamat"] = "No Data"
df = df[["nama", "alamat"]]
df
```

### b. Kabupaten Batubara

```python
url_1 = "https://setwan.batubarakab.go.id/fraksi/fraksi-partai-demokrasi-indonesia-perjuangan-pdip"
url_2 = "https://setwan.batubarakab.go.id/fraksi/fraksi-partai-golongan-karya-golkar"
url_3 = "https://setwan.batubarakab.go.id/fraksi/fraksi-partai-gerakan-indonesia-raya-gerindra"
url_4 = "https://setwan.batubarakab.go.id/fraksi/fraksi-partai-amanat-nasional-pan"
url_5 = "https://setwan.batubarakab.go.id/fraksi/fraksi-partai-keadilan-sejahtera-pks"
url_6 = "https://setwan.batubarakab.go.id/fraksi/fraksi-partai-demokrat"
url_7 = "https://setwan.batubarakab.go.id/fraksi/fraksi-partai-nasional-demokrat-nasdem"
url_8 = "https://setwan.batubarakab.go.id/fraksi/fraksi-partai-persatuan-pembangunan-ppp"
url_9 = "https://setwan.batubarakab.go.id/fraksi/fraksi-partai-bulan-bintang-pbb"
url_10 = "https://setwan.batubarakab.go.id/fraksi/fraksi-nurani-karya-bangsa-nkb"

urls = [url_1, url_2, url_3, url_4, url_5, url_6, url_7, url_8, url_9, url_10]
```

```python
list_nama = []
for url in urls:
    soup = get_url(url)
    spans = soup.find_all("span", {"class" : "thumb-info-inner"})
    for span in spans:
        list_nama.append(span.string)
```

```python
df = pd.DataFrame({"nama" : list_nama})
df["alamat"] = "No Data"
df
```

```python
df.to_csv("./DPRD TK II/Sumut/Batubara.csv", index=False)
```
