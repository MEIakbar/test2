```python
from bs4 import BeautifulSoup
import requests
from urllib.request import urlopen, urlretrieve, quote
from urllib.parse import urljoin
import pandas as pd
import csv
from tqdm import tqdm
import numpy as np
```

# Function

```python
def get_url(url):
    headers = {
         'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.142 Safari/537.36'
     }
    r = requests.get(url, headers=headers)  # Using the custom headers we defined above
    soup = BeautifulSoup(r.content, 'lxml') 
    return soup

def get_table(url):
    df = pd.read_html(url)[4]    
    df = df.rename(columns={"Nama Anggota" : "nama"})
    df["alamat"] = "No Data"
    df = df[["nama", "alamat"]]

    return df
```

### Kota Banjarmasin

```python
url_1 = "http://dprd.banjarmasinkota.go.id/pages/fraksi-pan"
url_2 = "http://dprd.banjarmasinkota.go.id/pages/fraksi-gerindra"
url_3 = "http://dprd.banjarmasinkota.go.id/pages/fraksi-golkar"
url_4 = "http://dprd.banjarmasinkota.go.id/pages/fraksi-pdi-p"
url_5 = "http://dprd.banjarmasinkota.go.id/pages/fraksi-pks"
url_6 = "http://dprd.banjarmasinkota.go.id/pages/fraksi-pkb"
url_7 = "http://dprd.banjarmasinkota.go.id/pages/fraksi-demokrat"
url_8 = "http://dprd.banjarmasinkota.go.id/pages/fraksi-bintang-restorasi"

urls = [url_1, url_2, url_3, url_4, url_5, url_6, url_7, url_8]
```

```python
li = []
for url in urls:
    df_list = pd.read_html(url)
    for idx in tqdm(range(len(df_list))):
        df = df_list[idx].T
        df.columns = ["nama", "ttl", "jabatan"]
        df = df.iloc[[1]]
        li.append(df)
    
```

```python
df = pd.concat(li, ignore_index=True)
df
```

```python
df["nama"] = [x.split(":")[1] for x in df["nama"]]
df["ttl"] = [x.split(":")[1] for x in df["ttl"]]
df["tempat lahir"] = [x.split(",")[0] for x in df["ttl"]]
df["tanggal lahir"] = [x.split(":")[1] if len(x.split(":")) == 2 else "No Data" for x in df["ttl"]]
df["alamat"] = "No Data"
df = df[["nama", "alamat", "tempat lahir", "tanggal lahir"]]
df
```

```python
df.to_csv("./DPRD TK II/Kalimantan Selatan/Banjarmasin.csv", index=False)
```

### Kota Samarinda

```python
url_1 = "https://dprd.samarindakota.go.id/dprd/alat-kelengkapan/fraksi/partai-demokrasi-indonesia-perjuangan"
url_2 = "https://dprd.samarindakota.go.id/dprd/alat-kelengkapan/fraksi/partai-gerakan-indonesia-raya"
url_3 = "https://dprd.samarindakota.go.id/dprd/alat-kelengkapan/fraksi/partai-golongan-karya"
url_4 = "https://dprd.samarindakota.go.id/dprd/alat-kelengkapan/fraksi/partai-keadilan-sejahtera"
url_5 = "https://dprd.samarindakota.go.id/dprd/alat-kelengkapan/fraksi/partai-demokrat"
url_6 = "https://dprd.samarindakota.go.id/dprd/alat-kelengkapan/fraksi/partai-nasdem"
url_7 = "https://dprd.samarindakota.go.id/dprd/alat-kelengkapan/fraksi/partai-amanat-nasional"
url_8 = "https://dprd.samarindakota.go.id/dprd/alat-kelengkapan/fraksi/partai-kebangkitan-pembangunan"


urls = [url_1, url_2, url_3, url_4, url_5, url_6, url_7, url_8]
```

```python
list_link = []

for url in urls:
    soup = get_url(url)
    list_link.extend([y["href"] for x in soup.find_all("div", {"class":"yellow struktur"}) for y in x.find_all("a", href=True)])
```

```python
li = []
for url in tqdm(list_link):
    soup = get_url(url)
    nama_string = soup.find("h2", {"style":"margin-bottom: 20px"}).text

    df = pd.read_html(url)[0].T
    df.columns = ["ttl", "agama", "istri", "alamat"]
    df = df.iloc[[2]]
    df["nama"] = nama_string
    df["tempat lahir"] = [x.split(",")[0] for x in df["ttl"]]
    df["tanggal lahir"] = [x.split(",")[1] if len(x.split(",")) >=2 else "No Data" for x in df["ttl"]]
    df = df[["nama", "alamat", "tempat lahir", "tanggal lahir"]]
    li.append(df)
df = pd.concat(li, ignore_index=True)
df
```

```python
df.to_csv("./DPRD TK II/Kalimantan Timur/Samarinda.csv", index=False)
```
