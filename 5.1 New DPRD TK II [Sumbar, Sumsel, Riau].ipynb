```python
from bs4 import BeautifulSoup
import requests
from urllib.request import urlopen, urlretrieve, quote
from urllib.parse import urljoin
import pandas as pd
import csv
from tqdm import tqdm
```

# 1. Function

```python
def get_url(url):
    headers = {
         'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.142 Safari/537.36'
     }
    r = requests.get(url, headers=headers)  # Using the custom headers we defined above
    soup = BeautifulSoup(r.content, 'lxml') 
    return soup

def get_table(url):
    df = pd.read_html(url)[4]    
    df = df.rename(columns={"Nama Anggota" : "nama"})
    df["alamat"] = "No Data"
    df = df[["nama", "alamat"]]

    return df

```

# 2. Sumsel

### Kabupaten Banyuasin


```python
url = "http://dprd.banyuasinkab.go.id/"
```

```python
soup = get_url(url)
```

```python
list_nama = []
spans = soup.find_all("a", {"class":"fancybox"}, href=True)
for span in spans:
    nama = str(span).split('title="')[1]
    list_nama.append(nama.split('"/><')[0])
list_nama
```

```python
df = pd.DataFrame.from_dict({"nama" : list_nama})
df["alamat"] = "No Data"
df
```

```python
df.to_csv("./DPRD TK II/Sumsel/Banyuasin.csv", index=False)
```

### Kabupaten Lahat


```python
url = "https://sumeks.co/ini-nama-nama-anggota-dprd-lahat-periode-2019-2024/"
soup = get_url(url)
```

```python
spans = soup.find_all("ol")
spans.find_all("li")
```

```python
list_nama = []
for span in spans:
    for child in span.find_all("li"):
        list_nama.append(child.text.split("/")[0])
```

```python
df.to_csv("./DPRD TK II/Sumsel/Lahat.csv", index=False)
```
