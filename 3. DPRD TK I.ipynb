```python
def get_lampung(url):
    headers = {
         'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.142 Safari/537.36'
     }
    r = requests.get(url, headers=headers)  # Using the custom headers we defined above
    soup = BeautifulSoup(r.content, 'html5lib') 
    return soup
```

```python
url_pdip = "https://setwan-dprd.lampungprov.go.id/pages/fraksi-pdi-p"
url_gerindra = "https://setwan-dprd.lampungprov.go.id/pages/fraksi-gerindra"
url_golkar = "https://setwan-dprd.lampungprov.go.id/pages/fraksi-golkar"
url_demokrat = "https://setwan-dprd.lampungprov.go.id/pages/fraksi-demokrat"
url_nasdem = "https://setwan-dprd.lampungprov.go.id/pages/fraksi-nasdem"
url_pkb = "https://setwan-dprd.lampungprov.go.id/pages/fraksi-pkb"
url_pks = "https://setwan-dprd.lampungprov.go.id/pages/fraksi-pks"
url_pan = "https://setwan-dprd.lampungprov.go.id/pages/fraksi-pan"

urls = [url_pdip, url_gerindra, url_golkar, url_demokrat, url_nasdem, url_pkb, url_pks, url_pan]
```

```python
list_link = []
for url in tqdm(urls):
    soup = get_lampung(url)
    spans = soup.find_all('span', {"style":"font-size:14px"})
    for s in spans:
        for x in s.find_all('a', href=True):
            list_link.append(x['href'])
```

```python
len(list_link)
```

```python
Nama_list = []
Deskripsi_list = []

for link in tqdm(list_link):
    temp_list = []
    Nama_list.append(link.split("/")[-1])
    soup = get_lampung(link)
    spans = soup.find_all('div', {"class":"single-post-text"})            
    try:
        for x in spans[0].find_all('p'):
            temp_list.append(x.string)
        Deskripsi_list.append(temp_list)
    except:
        Deskripsi_list.append("No Data")
        continue
```

```python
spans = soup.find_all('div', {"class":"single-post-text"})
spans
```

```python
def lapung_prepro(x):
    x = x.replace("<p>", "")
    x = x.replace("</p>", "")
    x = x.replace("<strong>", "")
    x = x.replace("</strong>", "")
    x = x.replace("\xa0", "")
    x = x.replace("h2", "")
    x = x.replace("/h2", "")
    x = x.replace('<a href="#">', "")
    x = x.replace("/a", "")
    x = x.replace("<>", "")
    x = x.replace("</>", "")
    return x
```

```python
x_list = []
x_failed = []
for link in tqdm(list_link):
    temp_list = []
    Nama_list.append(link.split("/")[-1])
    soup = get_lampung(link)
    spans = soup.find_all('div', {"class":"single-post-text"})
    try:
        for x in spans[0]:
            ele_text = str(x)
            if "\n" in ele_text:
                continue
            else:
                x = lapung_prepro(ele_text)
                temp_list.append(x)
    except:
        x_failed.append(link)
        continue
    x_list.append(temp_list)
```

```python
len(x_list), len(x_failed)
```

```python
x_list
```

```python
x_failed
```

```python
nama_list = []
ttl_list = []
for idx in x_list:
    nama_list.append(idx[0])
    ttl_list.append(idx[1])

for child in x_failed:
    nama_list.append(child.split("/")[-1].replace("-", " "))
    ttl_list.append("-")
```

```python
len(nama_list), len(ttl_list)
```

```python
df = pd.DataFrame.from_dict({"Nama" : nama_list,
                            "Tempat dan tanggal lahir" : ttl_list})
df
```

```python
df.to_csv("./All/DPRD TK I/Lampung.csv", index=False)
```

# Kalimantan Selatan

```python
url_pimpinan = "https://dprdkalselprov.id/pimpinan-dewan/"
url_1 = "https://dprdkalselprov.id/komisi-i/"
url_2 = "https://dprdkalselprov.id/komisi-ii/"
url_3 = "https://dprdkalselprov.id/komisi-iii/"
url_4 = "https://dprdkalselprov.id/komisi-iv/"

urls = [url_pimpinan, url_1, url_2, url_3, url_4]
```

```python
def get_data(url):
    headers = {
         'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.142 Safari/537.36'
     }
    r = requests.get(url, headers=headers)  # Using the custom headers we defined above
    soup = BeautifulSoup(r.content, 'html5lib') 
    return soup

def get_failed(url):
    soup = get_data(url)
    spans = soup.find_all('div', {"class":"wp-block-media-text__content"})
    ttl = spans[0].find_all("p", {"style":"font-size:15px"})[0].string
    split_ttl = ttl.split(":")[1]
    split_coma = split_ttl.split(",")
    tempat, tanggal = split_coma
    return tempat, tanggal
```

```python
list_link = []

for url in tqdm(urls):
    soup = get_data(url)
    spans = soup.find_all('div', {"class":"entry-content clearfix"})
    for x in spans[0].find_all('a', href=True):
        list_link.append(x['href'])    
```

```python
list_nama = []
list_tempat = []
list_tanggal = []
list_failed = []

for idx in tqdm(range(len(list_link))):
    url = list_link[idx]
    nama = url.split("/")[-2]   
    nama = nama.replace("-", " ")
    try:
        df_sample = pd.read_html(url)[0]
        split_coma = df_sample[1][0].split(",")
        tempat_lahir = split_coma[0]
        tangal_lahir = split_coma[1].split(" ")[1]

        list_nama.append(nama)
        list_tempat.append(tempat_lahir)
        list_tanggal.append(tangal_lahir)
    except:
        list_failed.append(url)
        
len(list_nama), len(list_tempat), len(list_tanggal), len(list_failed)
```

```python
failed_again = []
for idx in range(len(list_failed)):
    url = list_failed[idx]
    try:
        if idx == 0 or idx == 17:
            nama = url.split("/")[-1]
        else:
            nama = url.split("/")[-2]  
        nama = nama.replace("-", " ")
        tempat, tanggal = get_failed(url)

        list_nama.append(nama)
        list_tempat.append(tempat_lahir)
        list_tanggal.append(tangal_lahir)
    except:
        failed_again.append(url)
len(failed_again)
```

```python
len(list_nama), len(list_tempat), len(list_tanggal)
```

```python
df = pd.DataFrame.from_dict({"nama": list_nama,
                            "tempat lahir": list_tempat,
                            "tanggal lahir": list_tanggal})
```

```python
df.to_csv("./All/DPRD TK I/Kalimantan_Selatan.csv", index=False)
```

```python
url_1 = "https://dprd-dkijakartaprov.go.id/fraksi/fraksi-pdi-perjuangan/"
url_2 = "https://dprd-dkijakartaprov.go.id/fraksi/fraksi-partai-gerindra/"
url_3 = "https://dprd-dkijakartaprov.go.id/fraksi/fraksi-partai-keadilan-sejahtera/"
url_4 = "https://dprd-dkijakartaprov.go.id/fraksi/fraksi-partai-demokrat-partai-amanat-nasional/"
url_5 = "https://dprd-dkijakartaprov.go.id/fraksi/anggota-terpilih-partai-amanat-nasional/"
url_6 = "https://dprd-dkijakartaprov.go.id/fraksi/partai-solidaritas-indonesia/"
url_7 = "https://dprd-dkijakartaprov.go.id/fraksi/fraksi-partai-nasdem/"
url_8 = "https://dprd-dkijakartaprov.go.id/fraksi/fraksi-partai-golkar/"
url_9 = "https://dprd-dkijakartaprov.go.id/fraksi/fraksi-partai-kebangkitan-bangsa/"

Base_url = "https://dprd-dkijakartaprov.go.id"
urls = [url_1, url_2, url_3, url_4, url_5, url_6, url_7, url_8, url_9]
```

```python
list_link = []
for url in urls:
    soup = get_data(url)
    spans = soup.select('div', {"class":"innerpage-right"})
    for x in spans:
        for x2 in x.find_all('a', href=True):
            if "http" in x2['href']:
                continue
            elif x2['href'] == '/' or x2['href'] == '':
                continue
            else:
                list_link.append(Base_url + x2['href'])  
list_link = list(set(list_link))
```

```python
list_link
```

```python
spans
```

```python
list_nama = []
list_lahir = []

for idx in tqdm(range(len(list_link))):
    url = list_link[idx]
    nama = url.split("/")[-2].replace("-", " ")
    soup = get_data(url)
    spans = soup.find_all('div', {"class":"col-md-8"})
    for span in spans:
        lahir = span.find_all('p')[0].text.split("\n")[0].replace("\xa0", "")
    
    list_nama.append(nama)
    list_lahir.append(lahir)
```

```python
len(list_nama), len(list_lahir)
```

```python
df = pd.DataFrame.from_dict({"nama" : list_nama,
                            "list_lahir" : list_lahir})
df
```

```python
df.to_csv("./All/DPRD TK I/DKI_Jakarta.csv", index=False)
```
