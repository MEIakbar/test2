```python
from bs4 import BeautifulSoup
import requests
from urllib.request import urlopen, urlretrieve, quote
from urllib.parse import urljoin
import pandas as pd
import csv
from tqdm import tqdm
```

```python
def get_url(url):
    headers = {
         'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.142 Safari/537.36'
     }
    r = requests.get(url, headers=headers)  # Using the custom headers we defined above
    soup = BeautifulSoup(r.content, 'html5lib') 
    return soup
```

# 3. Bandung

```python
url = "https://dprd.bandung.go.id/profil/daftar-anggota"
```

```python
soup = get_url(url)
soup
```

```python
list_link = []
spans = soup.find_all('a', attrs={'class':'proling'}, href=True)
for span in spans:
    list_link.append(span['href'])
len(list_link)
```

```python
list_nama = []
list_tempat = []
list_tanggal = []
list_alamat = []

for url in tqdm(list_link):
    # get nama
    list_nama.append(url.split("/")[-1].replace("-", " "))
    soup = get_url(url)
    spans = soup.find_all('div', {'class':'a'})
    # spans
    for idx in range(len(spans)):
        if idx == 0:
            ttl = str(spans[idx]).replace('<div class="a">', '')
            ttl = ttl.replace("</span></div>", "")
            ttl = ttl.replace("<span>", "")
            ttl = ttl.replace("\xa0", "")

            tempat = ttl.split("<br/>")[0]
            tanggal = ttl.split("<br/>")[1]
            
            list_tempat.append(tempat)
            list_tanggal.append(tanggal)

        elif idx == 1:
            list_alamat.append(spans[idx].text)
```

```python
df = pd.DataFrame.from_dict({"Nama" : list_nama,
                            "Tempat Lahir" : list_tempat,
                            "Tanggal Lahir" : list_tanggal,
                            "Alamat" : list_alamat})
df
```

```python
df.to_csv("./All/Bandung.csv", index=False)
```

```python
def get_nama_list(soup, nama_list):
    spans = soup.find_all('h5')
    for span in spans:
        nama_list.append(span.string)
    return nama_list

# def get_df_list(url, df_list, link):
```

```python
nama_list = []
li = []

for url in tqdm(urls):
    soup = get_url(url)
    # get list of nama
    nama_list = get_nama_list(soup, nama_list)
    
    # get other information
    list_link = []
    spans = soup.find_all('div', {"class" : "bidang"})
    for span in spans:
        link_dpr = span.find_all("a", href=True)
        for idx in link_dpr:
            list_link.append(idx['href'])
    
    for lnk in list_link:
        df = pd.read_html(lnk)[0].T
        new_header = df.iloc[0]
        df = df[2:].reset_index(drop=True) 
        df.columns = new_header 
        li.append(df)
```
