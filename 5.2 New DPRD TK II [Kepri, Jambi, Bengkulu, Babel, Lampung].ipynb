# Function

```python
def get_url(url):
    headers = {
         'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.142 Safari/537.36'
     }
    r = requests.get(url, headers=headers)  # Using the custom headers we defined above
    soup = BeautifulSoup(r.content, 'lxml') 
    return soup

def get_table(url):
    df = pd.read_html(url)[4]    
    df = df.rename(columns={"Nama Anggota" : "nama"})
    df["alamat"] = "No Data"
    df = df[["nama", "alamat"]]

    return df

```

# Lampung

### Kabupaten Lampung Tengah


```python
base_url = "https://dprd.lampungtengahkab.go.id/"
url_1 = "https://dprd.lampungtengahkab.go.id/data-anggota?page=1"
url_2 = "https://dprd.lampungtengahkab.go.id/data-anggota?page=2"
url_3 = "https://dprd.lampungtengahkab.go.id/data-anggota?page=3"
url_4 = "https://dprd.lampungtengahkab.go.id/data-anggota?page=4"
url_5 = "https://dprd.lampungtengahkab.go.id/data-anggota?page=5"
url_6 = "https://dprd.lampungtengahkab.go.id/data-anggota?page=6"

urls = [url_1, url_2, url_3, url_4, url_5, url_6]
```

```python
list_link = []
for url in urls:
    soup = get_url(url)
    for span in soup.find_all("div", {"class":"col-lg-4 col-sm-6 mb-25"}):
        for child in span.find_all('a', href=True):
            list_link.append(child["href"])
```

```python
li = []
for url_child in list_link:
    url = base_url + url_child
    soup = get_url(url)
    
    list_nama = []
    soup = get_url(url)
    list_nama.append(soup.find_all("figcaption")[0].find("h3").text)
    
    try:
        df = pd.read_html(url)[0].T
        df.columns = df.iloc[1]
        df = df.iloc[[4]]
        df["nama"] = list_nama
        df = df.rename(columns={"Alamat" : "alamat"})
        df["tempat lahir"] = [x[0] for x in df["Tempat, Tanggal Lahir"].str.split(",")]
        df["tanggal lahir"] = [x[1] for x in df["Tempat, Tanggal Lahir"].str.split(",")]
        df = df[["nama", "alamat", "tempat lahir", "tanggal lahir"]].reset_index(drop=True)
    except:
        df = pd.DataFrame.from_dict({"nama" : list_nama})
        df["alamat"] = "No Data"
        df["tempat lahir"] = "No Data"
        df["tanggal lahir"] = "No Data"
    li.append(df)
```

```python
df = pd.concat(li, ignore_index=True)
```

```python
df.to_csv("./DPRD TK II/Lampung/Lampung_Tengah.csv", index=False)
```

### Kabupaten Lampung Barat


```python
base_url = "https://dprd-lampungbaratkab.go.id/"
# url = "https://dprd-lampungbaratkab.go.id/profil-dewan.html"
url_1 = "https://dprd-lampungbaratkab.go.id/profil-dewan-page1.html"
url_2 = "https://dprd-lampungbaratkab.go.id/profil-dewan-page2.html"
url_3 = "https://dprd-lampungbaratkab.go.id/profil-dewan-page3.html"
url_4 = "https://dprd-lampungbaratkab.go.id/profil-dewan-page4.html"

urls = [url_1, url_2, url_3, url_4]
```

```python
list_link = []
for url in urls:
    soup = get_url(url)
    for span in soup.find("div", {"class" : "isi"}).find_all("a", href=True):
        list_link.append(base_url + span["href"])
list_link = list(set(list_link))
```

```python
remove_ele = urls.copy()
remove_ele.append('https://dprd-lampungbaratkab.go.id/')
list_link = list(set(list_link) - set(remove_ele))
len(list_link), list_link
```

```python
li = []
for child_url in tqdm(list_link):
    df = pd.read_html(child_url)[0].T
    df.columns = df.iloc[1]
    df = df.iloc[[3]]
    
    df["tempat lahir"] = [x[0] for x in df["Tempat Tgl Lahir"].str.split(",")]
    df["tanggal lahir"] = [x[1] for x in df["Tempat Tgl Lahir"].str.split(",")]
    df = df.rename(columns={"Nama" : "nama",
                           "Alamat" : "alamat"})

    df = df[["nama", "alamat", "tempat lahir", "tanggal lahir"]].reset_index(drop=True)
    li.append(df)
```

```python
df = pd.concat(li, ignore_index=True)
df
```

```python
df.to_csv("./DPRD TK II/Lampung/Lampung_Barat.csv", index=False)
```

```python
url = "https://dprd.metrokota.go.id/anggota.html"
df_list = pd.read_html(url)
```

```python
li = []
for child in tqdm(df_list):
    df = child.T
    df.columns = df.iloc[1]
    df = df.rename(columns={"Nama" : "nama",
                           "Alamat" : "alamat"})
    df = df.iloc[[2]]
    df["tempat lahir"] = [x[0] for x in df["TTL"].str.split(",")]
    df["tanggal lahir"] = [x[1] for x in df["TTL"].str.split(",")]
    df = df[["nama", "alamat", "tempat lahir", "tanggal lahir"]].reset_index(drop=True)
    li.append(df)
```

```python
df = pd.concat(li, ignore_index=True)
df
```
